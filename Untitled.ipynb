```python
#Dependencies
import pandas as pd
import matplotlib.pyplot as plt
from config import user, pw 
from sqlalchemy import create_engine
```

```python
#import sql db
engine= create_engine(f'postgresql://{user}:{pw}@localhost:5432/PH_EmployeeDB')
connection=engine.connect()
```

```python
#Get salaries information
salaries=pd.read_sql('select * from salaries', connection)
salaries.head()
```

```python
#Get titles information
titles=pd.read_sql('select * from titles', connection)
titles.head()
```

```python
#get employees to connect salary and title
employees=pd.read_sql('select * from employees', connection)
employees.head()
```

```python
#task 1, create a histogram to visualize the most common salary ranges 
salaries.hist(column='salary')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('P.H. Employee Salary Frequency')
#plt.xlim(40000,125000)
#plt.ylim(0,160000)
plt.axis([40000,125000,0,160000])
plt.show()
```

```python
#merge tables together
emp_salary=salaries.merge(employees, on='emp_no')
emp_salary.head()
```

```python
#add titles
title_salary=emp_salary.merge(titles, left_on='emp_title_id',right_on='title_id')
title_salary.head()
```

```python
#task 2, create a bar chart of average salary by title
#first group into title and then average to get numbers to chart with
salary_chart=title_salary.groupby('title')['salary'].mean().round(2)
```

```python
#then chart
salary_chart.plot.bar()
plt.xlabel('Title')
plt.ylabel('Salary')
plt.title('Average Salary by Job Title')
plt.show()
```

```python
#close connection
connection.close()
```
